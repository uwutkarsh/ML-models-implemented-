```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Experience-Salary.csv")   
df.head()
```

```python
X= df["exp(in months)"].values.reshape(-1,1)
y = df["salary(in thousands)"].values.reshape(-1,1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
print("training successful")
```

```python
m = model.coef_[0]  #the slope 
b = model.intercept_  #the intercept
```

```python
#the linear regression line 
y_regre_line= model.predict(X)
plt.figure(figsize=(10,6))
plt.scatter(x=df["exp(in months)"],y=df["salary(in thousands)"] )
plt.show()
```
