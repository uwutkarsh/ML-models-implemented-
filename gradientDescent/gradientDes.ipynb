```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples= 500, n_features=1, n_informative= 1 , noise = 70)
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
class GradientDesc:
    def __init__(self,learning_rate,epochs):
            self.m  = -100
            self.b =  -200
            self.learning_rate = learning_rate
            self.epochs = epochs
    def fit(self, X,y):
          for i in range(self.epochs):
                loss_slopeB = -2*np.sum(y - self.m*X.ravel() - self.b)
                loss_slopeM = -2*np.sum(X.ravel()*(y - self.m*X.ravel() - self.b))
                self.b = self.b - self.learning_rate*loss_slopeB
                self.m = self.m - self.learning_rate*loss_slopeM
               
          print(self.m)
          print(self.b)

                
          
            
```

```python
GD = GradientDesc(0.3, 100)
GD.fit(X,y)
```
