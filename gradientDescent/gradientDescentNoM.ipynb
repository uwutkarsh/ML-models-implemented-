gradient descent for the B(intercept)

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=200, n_features=1,n_informative=1,n_targets=1,noise=50)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
b = reg.intercept_
print(b)

```

```python
class GradientDesLR:
    def __init__(self,learning_rate,epochs):
        self.m = reg.coef_    # taking teh actual value in this case
        self.b = -100 #taking any random value 
        self.lr = learning_rate
        self.epochs = epochs
    def fit(self,X,y):
        for i in range(self.epochs): 
            loss_slope = -2*np.sum(y - self.m*X.ravel()-self.b)
            self.b = self.b - (self.lr*loss_slope)
        print(self.b)

        
```

```python
GD = GradientDesLR(0.001,100)
```

```python
GD.fit(X,y)
```
